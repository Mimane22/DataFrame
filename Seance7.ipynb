```python
import pandas as pd
import numpy as np
```

```python
dict1 = {"name" : ["Anastasia", 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],

'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],

'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],

'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
data = pd.DataFrame(dict1, index = labels)
data
```

```python
# Print the three first rows
data.head(3)
```

```python
# Delete rows with Nan values
data = data.dropna()
data
```

```python
# Extract the 'name' and 'score' columns from the DataFrame
print(data[["name", "score"]])
```

```python
# append a new row 'k' to the DataFrame 

data.loc["k"] = ["Suresh", 15.5, 1, "yes"]
data
```

```python
# delete the 'attempts' column from the DataFrame

data = data.drop("attempts", axis = 1)
data
```

```python
# Add a new column "Success": if the score is higher than 10 we will have 1, else we will have 0

value = []
for i in data["score"]:
    if i > 10:
        value.append(1)
    else:
        value.append(0)

data['success'] = value
data
```

```python
# Export the DataFrame into a CSV file named "my_data"

data.to_csv('my_data.csv')
```
